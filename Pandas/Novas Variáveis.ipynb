## Criando novas variáveis
### Preparando o ambiente

```python
import pandas as pd
```

```python
alugueis_residenciais = pd.read_csv('dados/alugueis_residenciais_filtrados.csv', sep = ';')
alugueis_residenciais.sample(10)
```

### Tratando os dados
#### Valor Bruto

```python
alugueis_residenciais['Valor Bruto'] = alugueis_residenciais.Valor + alugueis_residenciais.Condominio + alugueis_residenciais.IPTU
```

```python
alugueis_residenciais
```

#### Valor do m²

```python
alugueis_residenciais['Valor m2'] = alugueis_residenciais.Valor / alugueis_residenciais.Area
alugueis_residenciais['Valor Bruto m2'] = alugueis_residenciais['Valor Bruto'] / alugueis_residenciais.Area

alugueis_residenciais['Valor m2'] = alugueis_residenciais['Valor m2'].round(2)
alugueis_residenciais['Valor Bruto m2'] = alugueis_residenciais['Valor Bruto m2'].round(2)
alugueis_residenciais
```

#### Agrupamento por tipos

```python
alugueis_residenciais.Tipo.drop_duplicates()
```

```python
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
#apartamento = ['Quitinete', 'Apartamento', 'Flat', 'Loft']
```

```python
alugueis_residenciais['Tipo Agregado'] = alugueis_residenciais['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
alugueis_residenciais.sample(10)
```
