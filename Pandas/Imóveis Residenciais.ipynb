## Imóveis residenciais
### Preparação do ambiente

```python
import pandas as pd
```

```python
alugueis = pd.read_csv('dados/aluguel.csv', sep=";")
alugueis.sample(10)
```

### Preparação dos dados

```python
lista_tipos = list(alugueis.Tipo.drop_duplicates())
lista_tipos
```

```python
residencial = ['Quitinete',
 'Casa',
 'Apartamento',
 'Casa de Condomínio',
 'Flat',
 'Casa de Vila',
 'Loft']
```

```python
e_residencial = alugueis.Tipo.isin(residencial)
```

```python
alugueis_residenciais = alugueis[e_residencial]
```

```python
alugueis_residenciais.sample(5)
```

```python
alugueis_residenciais.describe()
```

```python
list(alugueis_residenciais.Tipo.drop_duplicates())
```

```python
alugueis_residenciais.shape
```

```python
alugueis.shape
```

```python
alugueis.shape[0] - alugueis_residenciais.shape[0]
```

```python
alugueis_residenciais.index = range(alugueis_residenciais.shape[0])
alugueis_residenciais.head()
```

### Exportando os dados

```python
alugueis_residenciais.to_csv('dados/alugueis_residenciais.csv', sep = ';')
```

```python
imp_alugueis_residenciais = pd.read_csv('dados/alugueis_residenciais.csv', sep = ';')

# Foi exportado com o índice do DataFrame
imp_alugueis_residenciais
```

```python
# Para remover a exportação do índice automático, basta:
alugueis_residenciais.to_csv('dados/alugueis_residenciais.csv', sep = ';', index = False)

imp_alugueis_residenciais = pd.read_csv('dados/alugueis_residenciais.csv', sep = ';')

# Exportado agora sem os índices
imp_alugueis_residenciais
```
