### Preparação dos dados

```python
import pandas as pd

dados = pd.read_csv('dados/aluguel.csv', sep=";")
dados.sample(5)
```

### Manipulação dos dados

```python
dados.Tipo
```

```python
# Retorna uma Series
tipos_imovel = dados.Tipo

# Poderia ser substituido pelo unique
# O Inplace define que o array em questão será modificado
tipos_imovel.drop_duplicates(inplace=True)
tipos_imovel
```

### Organizando a visualização

```python
# Transformando a series em DataFrame
tipos_imovel = pd.DataFrame(tipos_imovel)
```
