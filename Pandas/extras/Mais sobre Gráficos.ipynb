```python
#Faz com que as plotagens do matplotlib sejam salvas dentro do próprio caderno
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (15, 8))
```

```python
dados = pd.read_csv('dados/aluguel.csv', sep = ';')
dados.head()
```

```python
area = plt.figure()
```

```python
g1 = area.add_subplot(2, 2, 1)
g2 = area.add_subplot(2, 2, 2)
g3 = area.add_subplot(2, 2, 3)
g4 = area.add_subplot(2, 2, 4)
```

```python
# Gráfico de disperção
g1.scatter(dados.Valor, dados.Area)
g1.set_title('Valor x Area')
area
```

```python
g2.hist(dados.Valor)
g2.set_title('Histograma de Valor')
area
```

```python
dados_g3 = dados.Valor.sample(100)
dados_g3.index = range(dados_g3.shape[0])
g3.plot(dados_g3)
g3.set_title('Amostra de dados')
area
```

```python
grupo = dados.groupby('Tipo')['Valor']
# index são os rótulos
g4.bar(grupo.mean().index, grupo.mean().values)
g4.set_title('Valor médio por tipo')
area
```

```python
area.savefig('dados/grafico.png', dpi = 600, bbox_inches = 'tight')
```
