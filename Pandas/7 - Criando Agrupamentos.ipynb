## Preparando o ambiente

```python
import pandas as pd
alugueis = pd.read_csv('dados/t_alugueis_residenciais.csv', sep = ';')
alugueis.sample(10)
```

## Obtendo variáveis

```python
media = alugueis.Valor.mean()
media
```

```python
alugueis.Bairro.unique()
```

```python
bairros = ['Barra da Tijuca', 'Copacabana', 'Ipanema', 'Leblon', 'Botafogo', 'Flamengo', 'Tijuca']
selecao = alugueis['Bairro'].isin(bairros)
alugueis_bairros_selecionados = alugueis[selecao]
alugueis_bairros_selecionados
```

```python
alugueis_bairros_selecionados['Bairro'].drop_duplicates()
```

```python
# Retorna um DataFrameGroupBy, um tipo de tupla, contendo o bairro e os respectivos dados do mesmo
grupo_bairro = alugueis_bairros_selecionados.groupby('Bairro')
```

```python
# Obter médias pelo loop
for bairro, dados in grupo_bairro:
    print(f'{bairro} -> {round(dados.Valor.mean(), 2)}')
```

```python
# Obter média pelo pandas (retorna uma Series)
grupo_bairro.Valor.mean().round(2)
```

```python
# Esta média pode ser feita com mais de um campo para o agrupamento (retorna um DataFrame)
grupo_bairro[["Valor", "Condominio"]].mean().round(2)
```

## Exercício

Utilizando o DataFrame dado abaixo, crie um DataFrame com as notas médias dos alunos, com duas casas decimais, segundo seu sexo.

```python
alunos = pd.DataFrame({'Nome': ['Ary', 'Cátia', 'Denis', 'Beto', 'Bruna', 'Dara', 'Carlos', 'Alice'], 
                        'Sexo': ['M', 'F', 'M', 'M', 'F', 'F', 'M', 'F'], 
                        'Idade': [15, 27, 56, 32, 42, 21, 19, 35], 
                        'Notas': [7.5, 2.5, 5.0, 10, 8.2, 7, 6, 5.6], 
                        'Aprovado': [True, False, False, True, True, True, False, False]}, 
                        columns = ['Nome', 'Idade', 'Sexo', 'Notas', 'Aprovado'])
alunos
```

```python
gb_alunos = alunos.groupby("Sexo")
```

```python
df_gp_alunos = pd.DataFrame(gb_alunos.Notas.mean().round(2))
df_gp_alunos.columns = ["Notas Médias"]
df_gp_alunos
```

#### Utilizando MatPlotLib para visualização dos dados
https://matplotlib.org/

```python
# Configura para que o gráfico seja exibido na tela
%matplotlib inline

# Importa o PyPlot do MatPlotLib
import matplotlib.pyplot as plt

# Configurando o tamanho padrão dos gráficos a serem utilizados.
# (O Collab nesse sentido parece ser bem melhor)
plt.rc('figure', figsize = (20, 10))
```

```python
fig = grupo_bairro.Valor.std().plot.bar(color = 'blue')
```

```python
fig = grupo_bairro['Valor'].mean().plot.bar(color = 'blue')
fig.set_ylabel('Valor do Aluguel')
fig.set_title('Valor Médio do Aluguel por Bairro', {'fontsize': 22})
```
