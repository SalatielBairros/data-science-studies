## Configurando o ambiente

```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.rc('figure', figsize = (15, 7))
```

## Carregando os dados

```python
notas = pd.read_csv('dados/ml-latest-small/ratings.csv')
filmes_filtrados = pd.read_csv('dados/ml-latest-small/filtred_movies.csv')
filmes = pd.read_csv('dados/ml-latest-small/movies.csv')
```

```python
filmes_filtrados.head()
```

```python
notas.columns = ['idUsuario', 'idFilme', 'nota', 'momento']
filmes.columns = ['idFilme', 'titulo', 'generos']
```

## Análise de dois filmes
Analisando o filme Toy Story (idFilme == 1) e Jumanji (idFilme == 2)

```python
notas_ts = notas.query('idFilme == 1')
notas_jm = notas.query('idFilme == 2')
```

```python
sns.boxplot(notas_ts.nota)
```

```python
sns.boxplot(notas_jm.nota)
```

```python
sns.boxplot(notas_jm.nota)
sns.boxplot(notas_ts.nota)
```

Note que ao imprimir dois, ficam sobrepostos.

```python
plt.boxplot([notas_jm.nota, notas_ts.nota], labels = ['Toy Story', 'Jumani'], showmeans = True, meanline = True)
```

Para sobrepor no boxplot, é necessário fazer:

```python
sns.boxplot(x = 'idFilme', y = 'nota', data = notas.query('idFilme in [1, 2]'))
```
