```python
import numpy as np
import pandas as pd
import math
from scipy import stats
```

```python
def estimativa_ponderada(p1, p2, n1, n2):
  return ((n1 * p1) + (n2 * p2)) / (n1 + n2)

def pvalue_comp_proporcoes(p1, p2, n1, n2):
  p = estimativa_ponderada(p1, p2, n1, n2)
  pp1 = (p * (1 - p)) / n1
  pp2 = (p * (1 - p)) / n2
  z = (p1 - p2) / math.sqrt(pp1 + pp2)
  return stats.norm.cdf(z * -1)
```

```python
colunas = ['Vacina', 'Geral', 'Morte']
dados = [
    ['AZ', 0.82, 0.9],
    ['CV', 0.5, 0.97],
    ['PF', 0.95, 0.8],
    ['JN', 0.72, 0.85],
    ['SP', 0.92, 0.85]
]

vacinas = pd.DataFrame(columns = colunas, data = dados)
vacinas['NaoGeral'] = 1 - vacinas.Geral
vacinas['NaoMorte'] = 1 - vacinas.Morte
vacinas['MorteComVacina'] = vacinas.NaoMorte * vacinas.NaoGeral

cv_morte_vacina = 0.015
amostra = 40000
pvalue = [float(f"{pvalue_comp_proporcoes(p, cv_morte_vacina, amostra, amostra):.9f}") for p in vacinas.MorteComVacina]
vacinas['pvalue'] = pvalue
vacinas['h_melhor_que_corona'] = vacinas.pvalue > 0.05
vacinas
```
