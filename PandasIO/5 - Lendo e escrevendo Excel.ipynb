# Lendo e escrevendo excel
## Preparando o ambiente

```python
import numpy as np
import pandas as pd
```

## Carregando os dados

```python
alunos = pd.read_json('dados/nomes.json')
matriculas_curso = pd.read_json('dados/matriculas_curso.json')
cursos = pd.read_json('dados/cursos.json')
matriculas = pd.read_json('dados/matriculas.json')
```

```python
cursos.index.name = 'id_curso'
cursos.reset_index()
```

## Nomes dos alunos da próxima turma de estatística básica

```python
cursos.query('id_curso == 16')
```

```python
id_estatistica = 16
turma_estatistica = matriculas.query('id_curso == 16').set_index('id_aluno').join(alunos['nome']).rename(columns = {'nome': 'Alunos do curso de ' + cursos.loc[id_estatistica]['nome_do_curso']})
turma_estatistica
```

```python
cursos.loc[id_estatistica]['nome_do_curso']
```

```python
turma_estatistica.index.name = 'id_aluno'
turma_estatistica
```

## Salvando dados

```python
turma_estatistica.to_excel('dados/turma_estatistica.xlsx', index = False)
```
