## Preparando o ambiente

```python
url = 'https://tabela-cursos.herokuapp.com/index.html'
```

## Carregando os dados

```python
cursos = pd.read_html(url)
cursos
```

```python
# Está sendo lido como List e não como DataFrame
type(cursos)
```

```python
# Ao chamar o HTMl, ele retorna uma lista de dataframes. Caso existam várias tabelas, retornarão vários dataframes.
type(cursos[0])
```

```python
cursos = cursos[0]
cursos
```

## Tratamento dos dados

```python
cursos.columns = ['nome_do_curso']
cursos
```

```python
cursos['id'] = cursos.index + 1
cursos.set_index('id', inplace = True)
cursos
```

## Salvando os dados

```python
cursos.to_json('dados/cursos.json')
pd.read_json('dados/cursos.json')
```
